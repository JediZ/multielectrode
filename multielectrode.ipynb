### 1. Loading and Exploring Data

#### 1.0 Loading Data

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from mymltools import *
%matplotlib inline

data = np.load('motor_dataset.npy')[()]
trialAngle = data['trial_angle'] 
trialAcq = data['trial_acq']
trialGo = data['trial_go']
trialMove = data['trial_move']
channelMap = data['channel_map']
spkChannels = data['spk_channels']
spkTimes = data['spk_times']
spkTemplates = data['spk_templates']
```

### 2. Gaussian Naive Bayes (GNB)

```python
y = trialAngle # the experimental results
threshold =15  # do not consider high firing rate neurons for now

selectedNeurons = basal<=threshold
N_selected = np.sum(selectedNeurons)
N_trials = np.shape(spkdata)[1]
NN = np.arange(np.shape(spkdata)[0])
NN = NN[selectedNeurons]

## select neurons
spkdata_selected = [[[0,0] for i in range(N_trials)] for n in range(N_selected)]
for i in range(N_trials):
   for n in range(N_selected):
       spkdata_selected[n][i][0]=spkdata[NN[n]][i][0]
       spkdata_selected[n][i][1]=spkdata[NN[n]][i][1]

basal_selected = basal[selectedNeurons]
spkChannels_selected = spkChannels[selectedNeurons]

## setup modeling data
## features chosen for modeling include: 
## spike frequency data (in G-M and M-T), basal firing rate, channels used for each selected neuron.
X = prepare_X(spkdata_selected,basal_selected,spkChannels_selected)    
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)

## now the model
clf= GaussianNB() 
clf.fit(X_train,y_train)    # fitting model
y_pred=clf.predict(X_test)  # prediction for test set
```

```python
plot_confusion_matrix(y_test,y_pred)
print metrics.classification_report(y_test,y_pred)
```
